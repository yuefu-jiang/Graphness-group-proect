```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.patches as mpatches


# ignore warning messages 
import sys
import warnings

%matplotlib inline

pd.set_option("display.max.columns", None)

if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

```python
df = pd.read_csv('adult.data', sep=", ", header=None)

df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
       'marital-status', 'occupation', 'relationship', 'race', 'sex',
       'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
       'salary-range']
```

```python
df.shape
```

```python
df = df[df["workclass"] != '?']
df = df[df["education"] != '?']
df = df[df["marital-status"] != '?']
df = df[df["occupation"] != '?']
df = df[df["relationship"] != '?']
df = df[df["race"] != '?']
df = df[df["sex"] != '?']
df = df[df["native-country"] != '?']
df.shape
```

```python
## Renaming columns for convenience

df.rename(columns = {'education-num': 'education rank', 'marital-status': 'marital status',
                        'capital-gain': 'capital gain', 'capital-loss': 'capital loss', 'hours-per-week': 'HPW',
                        'native-country': 'country', 'fnlwgt': 'Final Weight', 'salary-range': 'income'}, inplace = True)
```

```python
plt.figure(figsize = (10,10));
sns.catplot(y = 'occupation', hue = 'income', col = 'sex', kind = 'count', data = df, palette = 'magma');
```

```python
sns.set(rc={'figure.figsize':(9.7,6.27)})

sns.countplot(df['relationship'], palette='mako', hue='income', data=df).set_title('Relationship Influence on income');
```
