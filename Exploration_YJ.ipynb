```python
# data exploration on adult dataset
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
import seaborn as sns


df = pd.read_csv("adult.csv")
print(df)

# Pie charts on categorical variables
# race
df['Race'].value_counts().plot(kind='pie')
```

```python
# Relationship
df['Relationship'].value_counts().plot(kind='pie')
```

```python
# Sex
df['Relationship'].value_counts().plot(kind='pie')
```

```python
#Box plot on working hours per week between groups - if the income was less than 50k or not
df.boxplot(column='HoursPerWeek', by='IncomeLessThan50k', figsize=(7, 9))
```

```python
ax1 = df[df['Sex'] == 'Male'].plot(kind='scatter', x='HoursPerWeek', y='EducationNo', color='blue', alpha=0.3, figsize=(10, 7))
df[df['Sex'] == 'Female'].plot(kind='scatter', x='HoursPerWeek', y='EducationNo', color='red', alpha=0.3, figsize=(10 ,7), ax=ax1)
plt.legend(labels=['Males', 'Females'])
plt.title('Relationship between hours worked and Education length between sexes', size=24)
plt.xlabel('hours', size=18)
plt.ylabel('Yeas of Education', size=18);
```

```python
ax1 = df[df['IncomeLessThan50k'] == False].plot(kind='scatter', x='HoursPerWeek', y='FinalWeight', color='purple', alpha=0.3, figsize=(10, 7))
df[df['IncomeLessThan50k'] == True].plot(kind='scatter', x='HoursPerWeek', y='FinalWeight', color='yellow', alpha=0.3, figsize=(10 ,7), ax=ax1)
plt.legend(labels=['<50K', '>50K'])
plt.title('Relationship between hours worked and final weight', size=24)
plt.xlabel('hours', size=18)
plt.ylabel('final weight', size=18);
```

```python
ax1 = df[df['IncomeLessThan50k'] == False].plot(kind='scatter', x='Age', y='FinalWeight', color='orange', alpha=0.3, figsize=(10, 7))
df[df['IncomeLessThan50k'] == True].plot(kind='scatter', x='Age', y='FinalWeight', color='blue', alpha=0.3, figsize=(10 ,7), ax=ax1)
plt.legend(labels=['<50K', '>50K'])
plt.title('Relationship between hours worked and final weight', size=24)
plt.xlabel('hours', size=18)
plt.ylabel('final weight', size=18);
```
